```python
import os

# Define path to your videos
# We assume the structure is: data/videos_for_error_labeling/shot_name/video.mp4
PROJECT_ROOT = os.path.dirname(os.getcwd()) # Go up from 'notebooks'
BASE_DIR = os.path.join(PROJECT_ROOT, 'data', 'videos_for_error_labeling')

print(f"ðŸ“‚ Scanning: {BASE_DIR}")

# Loop through each folder (shot type)
for folder_name in os.listdir(BASE_DIR):
    folder_path = os.path.join(BASE_DIR, folder_name)
    
    # Check if it's actually a folder
    if os.path.isdir(folder_path):
        print(f"\nProcessing folder: {folder_name}...")
        
        files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.mp4', '.mov', '.avi'))]
        
        # Loop through files and rename them
        for index, filename in enumerate(files, 1):
            # Create new name: foldername_number.mp4
            # e.g., cover_drive_01.mp4
            extension = os.path.splitext(filename)[1]
            new_name = f"{folder_name}_{index:03d}{extension}"
            
            old_file = os.path.join(folder_path, filename)
            new_file = os.path.join(folder_path, new_name)
            
            # Rename
            if old_file != new_file:
                os.rename(old_file, new_file)
                print(f"  Renamed: {filename} -> {new_name}")
            else:
                print(f"  Skipping: {filename} (Already correct)")

print("\nâœ… All files renamed successfully!")
```
