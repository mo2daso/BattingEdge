```python
import os
import pandas as pd

# Path settings
PROJECT_ROOT = os.path.dirname(os.getcwd()) # Assumes running from 'notebooks'
VIDEO_DIR = os.path.join(PROJECT_ROOT, 'data', 'videos_for_error_labeling')
CSV_PATH = os.path.join(PROJECT_ROOT, 'data', 'error_labels.csv')

data_rows = []

print(f"Scanning: {VIDEO_DIR}")

# Walk through all folders
for root, dirs, files in os.walk(VIDEO_DIR):
    for file in files:
        if file.lower().endswith(('.mp4', '.mov', '.avi')):
            # Get info
            filename = file
            # Folder name is the shot type (e.g., 'cover_drive')
            shot_type = os.path.basename(root)
            
            data_rows.append({
                'video_filename': filename,
                'shot_type': shot_type,
                'error_label': '' # Empty for you to fill
            })

# Create DataFrame
df = pd.DataFrame(data_rows)

# Sort to make it easy to work with
df = df.sort_values(by=['shot_type', 'video_filename'])

# Save to CSV
df.to_csv(CSV_PATH, index=False)

print(f"âœ… Successfully created '{CSV_PATH}'")
print(f"ðŸ“‹ Total videos found: {len(df)}")
print("ðŸ‘‰ Now open this file in Excel and fill the 'error_label' column!")
```
